```python
#Importing required Libraries and Packages
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

```

```python
#Loading Meat Substitutes Sales Data for Year 2021-2021
MSub_21 = pd.read_excel('Fz_Rfg Substitute Meat_POS_2021.xlsx') #add your file path
MSub_22 = pd.read_excel('Fz_Rfg Substitute Meat_POS_2022.xlsx')
MSub_23 = pd.read_excel('Fz_Rfg Substitute Meat_POS_2023.xlsx')
```

```python
#concatenating above DataFrames into One Frame
framesMSub = [MSub_21,MSub_22,MSub_23]
M_Subs = pd.concat(framesMSub)
M_Subs.info()
```

```python
#Load The Product Attributes Sheet
P_Attributes = pd.read_excel('Product Attributes.xlsx')#add your file path
P_Attributes.info()
import pandas as pd

P_Attributes.head()
```

```python
# Checking for duplicate UPC codes
duplicates = P_Attributes[P_Attributes['UPC 13 digit'].duplicated(keep=False)]  # `keep=False` marks all duplicates

# Displaying duplicates
print("Duplicate UPC codes found in the dataset:")
print(duplicates[[ 'Product', 'UPC 13 digit']])

# Counting duplicates
print("\nNumber of duplicate entries:")
print(duplicates['UPC 13 digit'].value_counts())
```

```python
#Joining the Sales DataFrame with the Product Attributes DataFrame
Merged_Sub = pd.merge(M_Subs, P_Attributes, on='UPC 13 digit', how='left')
Merged_Sub.head(10)
```

```python
rename_dict = {
    'BEYOND BEEF': 'BEYOND',
    'BEYOND CHICKEN': 'BEYOND',
    'BEYOND MEAT': 'BEYOND',
    'BEYOND MEAT BEYOND BREAKFAST SAUSAGE': 'BEYOND',
    'BEYOND MEAT BEYOND BURGER': 'BEYOND',
    'BEYOND MEAT BEYOND SAUSAGE': 'BEYOND',
    'BEYOND MEAT COOKOUT CLASSIC': 'BEYOND',
    'BEYOND POPCORN CHICKEN': 'BEYOND',
    'BEYOND STACK BURGER': 'BEYOND',
    'BEYOND STEAK': 'BEYOND',
    'MORNINGSTAR FARMS CHIK NUGGETS': 'MORNINGSTAR FARMS',
    'MORNINGSTAR FARMS DISNEY INCOGMEATO': 'MORNINGSTAR FARMS',
    'MORNINGSTAR FARMS GRILLERS': 'MORNINGSTAR FARMS',
    'MORNINGSTAR FARMS GRILLERS PRIME': 'MORNINGSTAR FARMS',
    'MORNINGSTAR FARMS INCOGMEATO': 'MORNINGSTAR FARMS',
    'MORNINGSTAR FARMS MEAL STARTERS': 'MORNINGSTAR FARMS',
    'MORNINGSTAR FARMS VEGGIE MEAL STARTERS': 'MORNINGSTAR FARMS',
    'MORNINGSTAR FARMS VEGGITIZERS': 'MORNINGSTAR FARMS',
    'GARDEIN ULTIMATE PLANT BASED': 'GARDEIN',
    'LIGHTLIFE SMART BACON': 'LIGHTLIFE' ,
    'LIGHTLIFE SMART DELI': 'LIGHTLIFE',
       'LIGHTLIFE SMART DOGS': 'LIGHTLIFE', 
    'LIGHTLIFE SMART GROUND': 'LIGHTLIFE',
       'LIGHTLIFE SMART MENU': 'LIGHTLIFE', 
    'LIGHTLIFE SMART SAUSAGES': 'LIGHTLIFE',
    'LIGHTLIFE SMART TENDERS': 'LIGHTLIFE',
    'LIGHTLIFE TOFU PUPS': 'LIGHTLIFE',
    'LIGHTLIFE GIMME LEAN': 'LIGHTLIFE',
    'LIGHTLIFE SMART PATTIES': 'LIGHTLIFE'
}

# Assuming 'Brand Franchise Name' is the column you want to update
Merged_Sub['Brand Name'] = Merged_Sub['Brand Name'].replace(rename_dict)
```

```python
# Define a dictionary to rename the regions
region_rename_dict = {
    'Total US - Multi Outlet + Conv': 'Total US',
    'California - Standard - Multi Outlet + Conv': 'California',
    'Great Lakes - Standard - Multi Outlet + Conv': 'Great Lakes',
    'Mid-South - Standard - Multi Outlet + Conv': 'Mid-South',
    'Northeast - Standard - Multi Outlet + Conv': 'Northeast',
    'Plains - Standard - Multi Outlet + Conv': 'Plains',
    'South Central - Standard - Multi Outlet + Conv': 'South Central',
    'Southeast - Standard - Multi Outlet + Conv': 'Southeast',
    'West - Standard - Multi Outlet + Conv': 'West'
}

# Replace the old region names with the new ones
Merged_Sub['Geography'] = Merged_Sub['Geography'].replace(region_rename_dict)

# Print the modified DataFrame
Merged_Sub.head(10)
```

```python
Merged_Sub.info()
```

```python
#focus on meat substitutes
Merged_Final_df = Merged_Sub[Merged_Sub['Product Type'] == 'MEAT SUBSTITUTE']
```

```python
Merged_Final_df.head(10)
```

```python
Cols = ['UPC 13 digit','Brand Name','Product_x','Dollar Sales','Form','Flavor / Scent','Total Ounces','Package', 'Geography','Cooked Info','Price per Unit']


Final_df = Merged_Final_df[Cols]
```

```python
# Top 10 Brands 

# Aggregate dollar sales by brand name and sort in descending order
brand_sales = Final_df.groupby('Brand Name')['Dollar Sales'].sum().sort_values(ascending=False)

# Display the result
brand_sales.head(10)
```

```python
#TOTAL US

Total_US_df = Final_df[Final_df['Geography'] == 'Total US']

print("Data for Total US region:")
print(Total_US_df .head(5))
print(Total_US_df.shape)
```

```python

Final_df_C = california_df[Cols]
# Define aggregation rules
aggregation_rules = {
    'Dollar Sales': 'sum',
    'Total Ounces': 'first',
    'Geography': 'first',  
    'Product_x': 'first',
    'Package': 'first',
    'Form': 'first',
    'Flavor / Scent': 'first',
    'Brand Name' : 'first',
    'Cooked Info': 'first',
    'Price per Unit': 'first'
}

# Group by UPC code and aggregate according to the rules
Consolidated_df_C = Final_df_C.groupby('UPC 13 digit').agg(aggregation_rules)

# Reset index to turn the UPC 13 digit back into a column
Consolidated_df_C.reset_index(inplace=True)

# Display the consolidated DataFrame
Consolidated_df_C.head(10)
```
